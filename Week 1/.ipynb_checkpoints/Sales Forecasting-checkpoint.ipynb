# Day 1 :

 # <u> *Business Problem Understanding* <u/>
## <u> 1. Project Objective <u/>
* Predict next month's product-level sales for a retail chain to optimize inventory management and reduce stockouts/overstock situations.
## <u>2. Business Context<u/>
1. Why forecast?: Retailers typically lose 8-10% of sales due to stockouts and 10-15% due to overstocking

2. Use Case: Monthly purchase order planning, warehouse staffing, promotional planning

3. Impact: Every 1% improvement in forecast accuracy can reduce inventory costs by 2-3%

### <u>3. Scope & Constraints<u/>
1. Forecast Horizon: 1 month ahead

2. Frequency: Monthly predictions

3. Granularity: Product-level forecasting (may aggregate to category level if data sparse)

4. Time Frame: Use 2-3 years of historical data

5. Assumption: No major business disruptions (store openings/closures, pandemics)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import xgboost as xgb
import pickle
import os
```

```python
df = pd.read_csv('walmart.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
# Fix the date format (it's DD-MM-YYYY)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', dayfirst=True)
```

```python
df.isnull().sum() # No Null Values
```

### Converting weekly data to monthly

```python
# Create month column
df['YearMonth'] = df['Date'].dt.to_period('M')
```

```python
# Group by store and month
monthly_all = df.groupby(['Store', 'YearMonth']).agg({
    'Weekly_Sales': 'sum',
    'Temperature': 'mean',
    'Fuel_Price': 'mean',
    'CPI': 'mean',
    'Unemployment': 'mean',
    'Holiday_Flag': 'max'
}).reset_index()
```

```python
monthly_all = monthly_all.rename(columns={'Weekly_Sales': 'Monthly_Sales'})
monthly_all['Date'] = monthly_all['YearMonth'].dt.to_timestamp('M')
```

```python
monthly_all.head()
```

## Creating features

```python
# Sort by store and date
monthly_all = monthly_all.sort_values(['Store', 'Date']).reset_index(drop=True)
```

```python
# Time features
monthly_all['Month'] = monthly_all['Date'].dt.month
monthly_all['Year'] = monthly_all['Date'].dt.year
```

```python
# Create lag features per store
all_features_data = []
```

```python
# Process each store
for store_id in monthly_all['Store'].unique():
    store_df = monthly_all[monthly_all['Store'] == store_id].copy()
    
    # Sort by date
    store_df = store_df.sort_values('Date').reset_index(drop=True)
    
    # Lag features
    store_df['Prev_Month_Sales'] = store_df['Monthly_Sales'].shift(1)
    store_df['Prev_2_Months_Sales'] = store_df['Monthly_Sales'].shift(2)
    store_df['Prev_Year_Sales'] = store_df['Monthly_Sales'].shift(12)
    # Rolling averages
    store_df['Rolling_3M_Avg'] = store_df['Monthly_Sales'].rolling(window=3, min_periods=1).mean()
    store_df['Rolling_6M_Avg'] = store_df['Monthly_Sales'].rolling(window=6, min_periods=1).mean()
    
    # Month-over-month change
    store_df['MoM_Change'] = store_df['Monthly_Sales'].pct_change()
    
    # Target: Next month's sales
    store_df['Next_Month_Sales'] = store_df['Monthly_Sales'].shift(-1)
    
    all_features_data.append(store_df)
```

```python
# Combine all stores back
all_data = pd.concat(all_features_data, ignore_index=True)
```

```python
all_data.isnull().sum()
```

```python
# Drop rows with NaN (from lag features)
all_data = all_data.dropna()
```

## Preparing Data For Modeling

```python
features = ['Store', 'Month', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment',
            'Holiday_Flag', 'Prev_Month_Sales', 'Prev_2_Months_Sales',
            'Prev_Year_Sales', 'Rolling_3M_Avg', 'Rolling_6M_Avg', 'MoM_Change']
```

```python
X = all_data[features]
y = all_data['Next_Month_Sales']
```

```python
# Split data (time-based, keep stores together)
# First, get unique dates sorted
unique_dates = all_data['Date'].unique()
split_idx = int(len(unique_dates) * 0.8)
train_date_cutoff = unique_dates[split_idx]
```

```python
# Split by date
X_train = X[all_data['Date'] < train_date_cutoff]
X_test = X[all_data['Date'] >= train_date_cutoff]
y_train = y[all_data['Date'] < train_date_cutoff]
y_test = y[all_data['Date'] >= train_date_cutoff]
```

```python
# Scale features (excluding Store ID)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
feature_cols_to_scale = [f for f in features if f != 'Store']
```

```python
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
```

```python
X_train_scaled[feature_cols_to_scale] = scaler.fit_transform(X_train[feature_cols_to_scale])
X_test_scaled[feature_cols_to_scale] = scaler.transform(X_test[feature_cols_to_scale])
```

## Training XGBoost

```python
model_all = xgb.XGBRegressor(
    n_estimators=150,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)
```

```python
# Train the model
model_all.fit(X_train_scaled, y_train)
```

```python
# Make predictions
y_pred_train = model_all.predict(X_train_scaled)
y_pred_test = model_all.predict(X_test_scaled)
```

```python
# Calculate metrics
def calculate_metrics(y_true, y_pred, label="Dataset"):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    
    print(f"{label} Metrics:")
    print(f"  MAE: ${mae:,.2f}")
    print(f"  RMSE: ${rmse:,.2f}")
    print(f"  MAPE: {mape:.2f}%")
    print()
    
    return {'mae': mae, 'rmse': rmse, 'mape': mape}
```

```python
print("\n📈 Model Performance (ALL Stores):")
train_metrics = calculate_metrics(y_train, y_pred_train, "Training")
test_metrics = calculate_metrics(y_test, y_pred_test, "Testing")
```

## Training ARIMA models

```python
sample_stores = [1, 5, 10, 20]  # Sample stores to analyze
arima_results = {}
```

```python
from statsmodels.tsa.arima.model import ARIMA

for store_id in sample_stores:
    print(f"\nTraining ARIMA for Store {store_id}...")
    
    # Get store data
    store_data = all_data[all_data['Store'] == store_id].copy()
    store_series = store_data.set_index('Date')['Monthly_Sales']
    
    # Split for this store
    split_idx = int(len(store_series) * 0.8)
    train_store = store_series[:split_idx]
    test_store = store_series[split_idx:]
    
    try:
        # Simple ARIMA model
        arima_model = ARIMA(train_store, order=(1,1,1), seasonal_order=(1,1,1,12))
        arima_fit = arima_model.fit()
        
        # Forecast
        forecast = arima_fit.forecast(steps=len(test_store))
        
        # Calculate metrics
        mae = mean_absolute_error(test_store, forecast)
        rmse = np.sqrt(mean_squared_error(test_store, forecast))
        mape = np.mean(np.abs((test_store - forecast) / test_store)) * 100
        
        arima_results[store_id] = {
            'model': arima_fit,
            'metrics': {'mae': mae, 'rmse': rmse, 'mape': mape}
        }
        
        print(f"  Store {store_id} - MAPE: {mape:.2f}%")
        
    except Exception as e:
        print(f"  Store {store_id} - Error: {str(e)[:50]}...")
```

```python
# Calculate average ARIMA performance
if arima_results:
    avg_mape = np.mean([r['metrics']['mape'] for r in arima_results.values()])
    print(f"\n📊 Average ARIMA MAPE (sample stores): {avg_mape:.2f}%")
```

## Creating Visulaization

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Feature Importance
feature_importance = pd.DataFrame({
    'feature': features,
    'importance': model_all.feature_importances_
}).sort_values('importance', ascending=True)

axes[0, 0].barh(feature_importance['feature'], feature_importance['importance'])
axes[0, 0].set_xlabel('Importance')
axes[0, 0].set_title('Feature Importance (ALL Stores)')
axes[0, 0].grid(True, alpha=0.3)


# 2. Actual vs Predicted (sample of test data)
sample_size = min(100, len(y_test))
sample_idx = np.random.choice(len(y_test), sample_size, replace=False)

axes[0, 1].scatter(y_test.iloc[sample_idx], y_pred_test[sample_idx], alpha=0.6)
axes[0, 1].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0, 1].set_xlabel('Actual Sales ($)')
axes[0, 1].set_ylabel('Predicted Sales ($)')
axes[0, 1].set_title(f'Actual vs Predicted (Sample of {sample_size} points)')
axes[0, 1].grid(True, alpha=0.3)


# 3. Error by Store
store_errors = []
for store_id in X_test['Store'].unique():
    store_mask = X_test['Store'] == store_id
    if store_mask.sum() > 0:
        store_mape = np.mean(np.abs((y_test[store_mask] - y_pred_test[store_mask]) / y_test[store_mask])) * 100
        store_errors.append((store_id, store_mape))

store_errors_df = pd.DataFrame(store_errors, columns=['Store', 'MAPE']).sort_values('MAPE').head(10)
axes[1, 0].bar(range(len(store_errors_df)), store_errors_df['MAPE'])
axes[1, 0].set_xticks(range(len(store_errors_df)))
axes[1, 0].set_xticklabels(store_errors_df['Store'])
axes[1, 0].set_xlabel('Store ID')
axes[1, 0].set_ylabel('MAPE (%)')
axes[1, 0].set_title('Top 10 Stores with Lowest Error')
axes[1, 0].grid(True, alpha=0.3)

# 4. Model Comparison
if arima_results:
    xgboost_avg_mape = test_metrics['mape']
    arima_avg_mape = avg_mape
    
    models = ['XGBoost (All Stores)', 'ARIMA (Avg Sample)']
    mape_values = [xgboost_avg_mape, arima_avg_mape]
    colors = ['blue', 'green']
    
    bars = axes[1, 1].bar(models, mape_values, color=colors)
    axes[1, 1].set_ylabel('MAPE (%)')
    axes[1, 1].set_title('Model Comparison (Lower is better)')
    axes[1, 1].grid(True, alpha=0.3)
    
    for bar, value in zip(bars, mape_values):
        height = bar.get_height()
        axes[1, 1].text(bar.get_x() + bar.get_width()/2, height + 0.5,
                       f'{value:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()

```

```python
print("\n🔮 Creating prediction function for ANY store...")

def predict_sales(store_id, month, temperature, fuel_price, cpi, unemployment,
                  holiday_flag, prev_month_sales, prev_2_months_sales,
                  prev_year_sales, model_type='xgboost'):
    """
    Predict next month's sales for any store
    
    Parameters:
    store_id: Store ID (1-45)
    month: Month (1-12)
    Other features as per training
    model_type: 'xgboost' or 'arima'
    """
    
    if model_type == 'xgboost':
        try:
            # Load model and scaler
            with open('models/xgboost_all_stores.pkl', 'rb') as f:
                model = pickle.load(f)
            
            with open('models/scaler_all_stores.pkl', 'rb') as f:
                scaler = pickle.load(f)
            
            with open('models/features_all_stores.pkl', 'rb') as f:
                features_list = pickle.load(f)
            
            # Create input DataFrame
            input_data = {
                'Store': store_id,
                'Month': month,
                'Temperature': temperature,
                'Fuel_Price': fuel_price,
                'CPI': cpi,
                'Unemployment': unemployment,
                'Holiday_Flag': holiday_flag,
                'Prev_Month_Sales': prev_month_sales,
                'Prev_2_Months_Sales': prev_2_months_sales,
                'Prev_Year_Sales': prev_year_sales,
                'Rolling_3M_Avg': (prev_month_sales + prev_2_months_sales) / 2,
                'Rolling_6M_Avg': (prev_month_sales + prev_2_months_sales) / 2,
                'MoM_Change': (prev_month_sales - prev_2_months_sales) / prev_2_months_sales if prev_2_months_sales != 0 else 0
            }
            
            # Create DataFrame
            input_df = pd.DataFrame([input_data])
            
            # Scale features
            feature_cols_to_scale = [f for f in features_list if f != 'Store']
            input_df_scaled = input_df.copy()
            input_df_scaled[feature_cols_to_scale] = scaler.transform(input_df[feature_cols_to_scale])
            
            # Make prediction
            prediction = model.predict(input_df_scaled[features_list])[0]
            
            return round(prediction, 2)
            
        except Exception as e:
            print(f"Error with XGBoost prediction: {e}")
            return None
    
    elif model_type == 'arima':
        try:
            # Try to load ARIMA model for this specific store
            arima_file = f'models/arima_store_{store_id}.pkl'
            if os.path.exists(arima_file):
                with open(arima_file, 'rb') as f:
                    arima_model = pickle.load(f)
                
                # ARIMA forecast (needs historical data)
                forecast = arima_model.forecast(steps=1)
                return round(forecast.iloc[0], 2)
            else:
                print(f"No ARIMA model found for Store {store_id}")
                return None
                
        except Exception as e:
            print(f"Error with ARIMA prediction: {e}")
            return None
    
    else:
        print("Error: model_type must be 'xgboost' or 'arima'")
        return None
```

```python
# Test prediction
print("\n🧪 Testing prediction for Store 1:")
prediction = predict_sales(
    store_id=1,
    month=12,
    temperature=65.0,
    fuel_price=3.50,
    cpi=220.0,
    unemployment=5.0,
    holiday_flag=1,
    prev_month_sales=250000,
    prev_2_months_sales=240000,
    prev_year_sales=280000,
    model_type='xgboost'
)
```
